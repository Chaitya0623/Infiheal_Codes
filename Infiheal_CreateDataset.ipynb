#Prompt

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from langchain import PromptTemplate

generate_example_template = """

% INSTRUCTIONS
You are a person suffering from procrastination.

% TEXTUAL QUESTION
{label}

% YOUR TASK
Generate an example in first person perspective that a person can face for the given label
"""
```

```python
import pandas as pd
from langchain.chains import LLMChain

df = pd.DataFrame(columns=['labels','text'])

for i in range(len(procrastination_model_labels)):
  for _ in range(10):
    label = procrastination_model_labels[i]
    prompt = PromptTemplate.from_template(generate_example_template)
    formatted_prompt = prompt.format(label=label)

    chain = LLMChain(llm=llm, prompt=prompt)
    example = chain.run(formatted_prompt)
    example = example.replace('\n','')
    new_row = {'labels': label,'text': example}
    df = df.append(new_row, ignore_index=True)
```
