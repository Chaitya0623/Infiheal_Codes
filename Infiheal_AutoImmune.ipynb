```python
import pandas as pd
df = pd.DataFrame(columns=['labels','text'])
```

```python
import torch
from transformers import pipeline

generate_text = pipeline(model="databricks/dolly-v2-3b", torch_dtype=torch.bfloat16, trust_remote_code=True, device_map="auto", device=0)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
health_conditions = [
    "Blood pressure",
    "Gut Issues (IBS, constipation, diarrhoea, bloating, belching, stomach upset)",
    "Diabetes",
    "Thyroid disease (Hyperthyroidism, Hypothyroidism)",
    "Cancer",
    "Autoimmune Disorders (Lupus, Connective Issue Syndrome)",
    "Spinal Injuries (paraplegia, Complete Tetraplegia (Quadriplegia))",
    "Headaches (Migraines, Tension Headaches)",
    "Chronic Pain",
    "Skin Issues (Eczema, psoriasis, vitiligo etc.)",
    "Heart Disease",
    "Renal disease",
    "Liver disease",
    "Respiratory issues (Asthma, COPD, Cystic Fibrosis)",
    "Urinary Issues",
    "Reproductive issues (infertility)",
    "Tinnitus",
    "Vascular Illnesses",
    "Vestibular Balance Disorder",
    "Acromegaly",
    "Cushing's Syndrome",
    "Cystic Fibrosis",
    "Grave's Disease",
    "Hashimoto's disease",
    "Ehlers-Danlos syndrome (EDS)",
    "Sexual dysfunction (vaginismus, erectile dysfunction etc)",
]
```

```python
filepath = '/content/drive/MyDrive/Internships/Infiheal/Self_Issues_Model.xlsx'
```

```python
import gc

for i in range(len(health_conditions)):
  gc.collect()
  for _ in range(100):
    label=health_conditions[i]

    res = generate_text(f"I want to share my story about me currently facing self issues because of {label}.")
    new_row = {'labels': label,'text': res[0]['generated_text'].replace('\n','')}
    df = df.append(new_row, ignore_index=True)
  df.to_excel(filepath, index=False)
  print(i)
```
